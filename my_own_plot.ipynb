# Plotting with matplotlib

```python
import numpy as np
import matplotlib
```

```python
matplotlib.use('MacOSX')
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
```

# Create a nice plot

Creating data to use in plot

```python
candy = ["Smarties", "Mentos", "Skittles", "Kit Kat",
              "M&M", "Twizzlers", "Marabou"]
students = ["Jessica", "Amanda", "Jonathan",
           "Omar", "Simon", "Nina", "Sarah"]

amount = np.array([[0.8, 2.4, 2.5, 3.9, 0.0, 4.0, 0.6],
                    [2.4, 0.0, 5.9, 1.0, 2.7, 0.0, 0.0],
                    [1.1, 2.4, 0.8, 4.3, 1.9, 4.4, 0.0],
                    [0.6, 0.0, 0.3, 0.0, 3.1, 0.0, 0.0],
                    [0.7, 1.7, 0.6, 2.6, 2.2, 5.2, 0.0],
                    [1.3, 1.2, 0.0, 0.0, 0.0, 3.2, 4.1],
                    [3.1, 4.0, 1.5, 5.4, 0.6, 1.9, 6.3]])
```

Plot results

```python
fig, ax = plt.subplots()
im = ax.imshow(amount, cmap = "coolwarm")

#Label x and y axis with respective list entries
ax.set_xticks(np.arange(len(students)), labels=students)
ax.set_yticks(np.arange(len(candy)), labels=candy)

#Rotate x axis label
plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")

#Loop over data
for i in range(len(candy)):
    for j in range(len(students)):
        text = ax.text(j, i, amount[i, j],
                       ha="center", va="center", color="w")
        
#Set title and plot results
ax.set_title("Candy consumption among students (in kg/year)", fontweight="bold", pad = 7.5)
fig.tight_layout()
plt.show()
```
